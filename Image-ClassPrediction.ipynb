## train AP

```python
train_AP =  pd.read_csv('/content/drive/MyDrive/kaggle/train/train_AP.csv')
truth = pd.read_csv('/content/BUU-LSPINE_spondylolisthesis/train/train_ground_truth.csv')

# Create a new column in XX dataframe with 'id' format of AP dataframe
truth['new_id'] = truth['id'].apply(lambda x: x.split('_')[0])

# Merge the dataframes
merged_df = pd.merge(train_AP, truth, how='left', left_on='id', right_on='new_id')

# If you want to drop the 'new_id' column after merge
merged_df = merged_df.drop(columns=['Unnamed: 0','new_id'])
merged_df['L'] = merged_df['id_y'].apply(lambda x : x.split('_')[-1])
train_AP_finish = merged_df.drop(columns=['id_y'])
train_AP_finish = train_AP_finish.rename(columns={'id_x':'id'})
train_AP_finish.head(),len(train_AP_finish)
```

```python
train_AP_finish['class'] =train_AP_finish['class'].astype(str)
```

```python
from tqdm.notebook import tqdm
import numpy as np
```

```python
new_rows = []

# Iterate over rows
for index, row in tqdm(train_AP_finish.iterrows()):
    # Check if 'class' is not equal to '0'
    if row['class'] != '5':
        # Create 5 new rows for each row with 'class' not equal to '0'
        for _ in range(1):
            new_row = row.copy()
            # Iterate over each column
            # Add constant value between (-0.1) and (0.1)
            constant_value = np.random.uniform(-0.1, 0.1)
            for column in train_AP_finish.columns:
                if column not in ['id', 'class','L']:
                    # Multiply value between 0.95 and 1.05
                    new_row[column] = new_row[column] + constant_value*np.random.uniform(0.95, 1.05)
            new_rows.append(new_row)

# Append the new rows to the dataframe
train_AP_finish = train_AP_finish.append(new_rows, ignore_index=True)
```

## train LA

```python
train_LA =  pd.read_csv('/content/drive/MyDrive/kaggle/train/train_LA.csv')
truth = pd.read_csv('/content/BUU-LSPINE_spondylolisthesis/train/train_ground_truth.csv')
# Create a new column in XX dataframe with 'id' format of AP dataframe
truth['new_id'] = truth['id'].apply(lambda x: x.split('_')[0])

# Merge the dataframes
merged_df = pd.merge(train_LA, truth, how='left', left_on='id', right_on='new_id')

# If you want to drop the 'new_id' column after merge
merged_df = merged_df.drop(columns=['Unnamed: 0','new_id'])
merged_df['L'] = merged_df['id_y'].apply(lambda x : x.split('_')[-1])
merged_df = merged_df.drop(columns=['id_y'])
merged_df['class'] =  merged_df['class'].astype(str)
train_LA_finish = merged_df.copy()
train_LA_finish = train_LA_finish.rename(columns={'id_x':'id'})
train_LA_finish.head()
```

```python
new_rows = []

# Iterate over rows
for index, row in tqdm(train_LA_finish.iterrows()):
    # Check if 'class' is not equal to '0'
    if row['class'] != '0':
        # Create 5 new rows for each row with 'class' not equal to '0'
        for _ in range(50):
            new_row = row.copy()
            # Add constant value between (-0.1) and (0.1)
            constant_value = np.random.uniform(-0.1, 0.1)
            # Iterate over each column
            for column in train_LA_finish.columns:
                if column not in ['id', 'class','L']:

                    # Multiply value between 0.95 and 1.05
                    new_row[column] = new_row[column] + constant_value* np.random.uniform(0.95, 1.05)

            new_rows.append(new_row)

# Append the new rows to the dataframe
train_LA_finish = train_LA_finish.append(new_rows, ignore_index=True)
```
